```python
import cv2
```

```python
def img_to_vid(img, vid_file='output_video.mp4'):
    height, width, channels = img.shape

    # Define video parameters
    fps = 30
    duration = 5  # seconds
    frame_count = fps * duration
    output_path = vid_file

    # Define the video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Write the image multiple times
    for _ in range(frame_count):
        out.write(img)

    out.release()
```

```python
left_img = cv2.imread('data/left_street.jpeg')
right_img = cv2.imread('data/right_street.jpeg')
img_to_vid(left_img, "left_street_vid.mp4")
img_to_vid(right_img, "right_street_vid.mp4")
```
